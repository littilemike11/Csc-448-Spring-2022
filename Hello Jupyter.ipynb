````python
import random
# get input from user
start = int(input("Input starting Value:"))

def make_series(head):
    series = [head]
    for i in range(1, 1000):
        r = random.randint(0, 10000)
        series.append(r)
    # print(series)
    return series

def find_distance(val1, val2):
    return abs(val2 - val1)

def fcfs(series):
    tempdistance = 0
    totaldistance = 0
    for i in range(0, 999):
        tempdistance = find_distance(series[i], series[i + 1])
        totaldistance += tempdistance
    print("FCFS Distance:", totaldistance)
    return totaldistance

def scan(series):
# get list of numbers < head
    head = series[0]
    less_than_result = filter(lambda x: x < head, series)
    list_less_than_head = list(less_than_result)
    list_less_than_head.sort(reverse=True) # reverse list of # <= head to minimize distance
    # print(list_less_than_head)
    # get list of numbers > head
    more_than_result = filter(lambda x: x > head, series)
    list_more_than_head = list(more_than_result)
    list_more_than_head.sort()
    # print(list_more_than_head)
    totaldistance = 0
    less_than_len = len(list_less_than_head)
    more_than_len = len(list_more_than_head)
    if head <= 5000: # if head <= half go down 1st
        if less_than_len != 0: # if no values less than head, go straight up
            less_last = list_less_than_head[-1]
            totaldistance += find_distance(head, list_less_than_head[0]) # go from head to 1st elem of less than list
            if less_than_len >= 2:
                for i in range(0, less_than_len - 1):
                    totaldistance += find_distance(list_less_than_head[i], list_less_than_head[i + 1])
            totaldistance += find_distance(less_last, 0) # go down to 0

        if more_than_len != 0:
            totaldistance += find_distance(0, list_more_than_head[0]) # go from 0 to values > head
            if more_than_len >= 2:
                for i in range(0, more_than_len - 1): # go through values more than head
                    totaldistance += find_distance(list_more_than_head[i], list_more_than_head[i + 1])
        print("scan less distance:", totaldistance)
        return totaldistance
    else: # if head >= half, go up 1st
        if more_than_len != 0:
            more_last = list_more_than_head[-1] # get last element of more than head list
            totaldistance += find_distance(head, list_more_than_head[0])
            if more_than_len >= 2:
                for i in range(0, more_than_len - 1): # go through values > head
                    totaldistance += find_distance(list_more_than_head[i], list_more_than_head[i + 1])
            totaldistance += find_distance(more_last, 9999) # get distance from last element of more than to max
        
        if less_than_len != 0:
            totaldistance += find_distance(9999,list_less_than_head[0]) # get distance from max to start of less than list
            if less_than_len >= 2:
                for i in range(0, less_than_len - 1):
                    totaldistance += find_distance(list_less_than_head[i], list_less_than_head[i + 1])
        print("scan More distance:", totaldistance)
        return totaldistance
    
def cscan(series):
    totaldistance = 0
    # get list of numbers < head
    head = series[0]
    less_than_result = filter(lambda x: x < head, series)
    list_less_than_head = list(less_than_result)
    list_less_than_head.sort()
    #print(list_less_than_head)
    # get list of numbers > head
    more_than_result = filter(lambda x: x > head, series)
    list_more_than_head = list(more_than_result)
    list_more_than_head.sort()
    # print(list_more_than_head)
    less_than_len = len(list_less_than_head)
    more_than_len = len(list_more_than_head)
    
    if less_than_len == 0: # if no less than values, only go up
        totaldistance += find_distance(head, list_more_than_head[0])
        for i in range(0, more_than_len - 1):
            totaldistance += find_distance(list_more_than_head[i], list_more_than_head[i + 1])
    
    if more_than_len == 0: # if no more than values, only go down
        totaldistance += find_distance(head, list_less_than_head[-1])
        for i in range(0, less_than_len - 1):
            totaldistance += find_distance(list_less_than_head[i], list_less_than_head[i + 1])

    if more_than_len != 0 and less_than_len != 0: # if neither or empty
        totaldistance += find_distance(head, list_more_than_head[0]) # go from head to values > head
        if more_than_len >= 2:
            for i in range(0, more_than_len - 1): # go through values > head
                totaldistance += find_distance(list_more_than_head[i], list_more_than_head[i + 1])
        totaldistance += find_distance(list_more_than_head[-1], 9999) # go from last elem of >head to max
        totaldistance += find_distance(0, 9999) # go back to 0 w/o checking any values
        totaldistance += find_distance(0, list_less_than_head[0]) # go from 0 to start of values< head
        if less_than_len >= 2:
            for i in range(0, less_than_len - 1): # go through values < head
                totaldistance += find_distance(list_less_than_head[i], list_less_than_head[i + 1])
    print("cscan Distance: ", totaldistance)
    return totaldistance

series = make_series(start)
fcfs_result = fcfs(series)
scan(series)
cscan(series)
````

Write a program that implements the following disk-scheduling algorithms:
* FCFS
* SCAN
* C-SCAN

Program will service a disk with 10,000 cylinders numbered 0 to 9,999. The program will
generate a random series of 1,000 cylinders requests and service them according to each of the
algorithms listed above. The program will be passed the initial position of the disk head (as a
parameter on the command line), and report the total amount of head movement required by
each algorithm.

First Come First Serve ( FSCS)- The cylinders are serviced in the order that they arrive into the
algorithm.
For example the series [15, 28 , 10, 9, 17,2], where 5 is the head will get serviced in the order 5,
28, 10, 9 ,17,2.

SCAN- SCAN algorithm also called the elevator algorithm works similar to an elevator. Given a
head, it moves towards one direction of the disk ( up or down), servicing all the cylinders that are
along the way, until it reaches the end of the disk. Then it reverses, or goes in the opposite
direction and services all the cylinders in the new direction.
For example the series [15, 28 , 10, 9, 17,2] where 15 is the head will be serviced 15, 17, 28, 30,
10, 9, 2. Note: 30 is the end of the disk.

C-SCAN- C-SCAN is similar to the scan algorithm. C-SCAN will move in one direction, servicing
all cylinders along the way until it reaches the end. Now it will go directly to the opposite end of
the disk, skipping any cylinders along the way, then continue from the start to the other cylinders
in the original direction it was travelling.
For example the series [15, 28 , 10, 9, 17, 2] where 15 is head will be serviced. 15, 17, 28, 30, 0, 2,
9, 10. Note 0 and 30 are the ends of the disks.

I believe SCAN is most efficient. FCFS is not efficient because it results in a lot of back and forth
servicing, which will increase the distance. SCAN and CSCAN are more efficient as they work in
sorted sublists and one move in one direction at a time. However, because CSCAN goes from
one end of the disk to the other end of the disk, it travels the full length of the disc at least once.
Therefore SCAN is most efficient because it minimizes the distance the head has to travel
